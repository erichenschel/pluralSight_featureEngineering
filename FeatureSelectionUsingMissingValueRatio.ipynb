```python
melb_data[['BuildingArea', 'YearBuilt', 'Landsize', 'Bathroom', 'Bedroom2', 'Car', 'Price', 'Longtitude', 'Lattitude']].isnull().sum()
```

```python
melb_data[['BuildingArea', 'YearBuilt', 'Landsize', 'Bathroom', 'Bedroom2', 'Car', 'Price', 'Longtitude', 'Lattitude']].describe()
```

```python
melb_data['Car'] = melb_data['Car'].fillna(melb_data['Car'].median())
melb_data['Price'] = melb_data['Price'].fillna(melb_data['Price'].median())
melb_data[['BuildingArea', 'YearBuilt', 'Landsize', 'Bathroom', 'Bedroom2', 'Car', 'Price', 'Longtitude', 'Lattitude']].describe()
```

```python
melb_data[['BuildingArea', 'YearBuilt', 'Landsize', 'Bathroom', 'Bedroom2', 'Car', 'Price', 'Longtitude', 'Lattitude']].isnull().sum()
```

```python
melb_data.columns
```

```python
melb_data[['BuildingArea', 'YearBuilt', 'Landsize', 'Bathroom', 'Bedroom2', 'Car', 'Price', 'Longtitude', 'Lattitude']].tail()
```

```python
melb_data['CouncilArea'].mode()
```

```python
melb_data['CouncilArea'] = melb_data['CouncilArea']\
.fillna(melb_data['CouncilArea'].mode()[0])

melb_data['Distance'] = melb_data['Distance']\
.fillna(melb_data['Distance'].mode()[0])

melb_data['Postcode'] = melb_data['Postcode']\
.fillna(melb_data['Postcode'].mode()[0])

melb_data['Regionname'] = melb_data['Regionname']\
.fillna(melb_data['Regionname'].mode()[0])

melb_data['Propertycount'] = melb_data['Propertycount']\
.fillna(melb_data['Propertycount'].mode()[0])
```

```python
melb_data.dropna(thresh=int(melb_data.shape[0] * 0.75), axis=1, inplace=True)
```
